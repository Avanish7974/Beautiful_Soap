```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.shopsy.in/search?q=mobiles+under+5000&as=on&as-show=on&p[]=facets.price_range.from=Min&p[]=facets.price_range.to%3DMax"
```

```python
r = requests.get(url)
print(r)  # to check the status code of the request
# respone = 200  # if the status code is 200, it means the request was successful
```

```python
soup = BeautifulSoup(r.text,"html.parser")
print(soup)
 # to print the HTML content of the page in a formatted way
```
